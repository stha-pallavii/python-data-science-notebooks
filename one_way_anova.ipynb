```python
# group the data by the column "Group"
grouped_df = medication_df.groupby("Group")
```

### ANOVA

```python
k = len(grouped_df.nunique()) #no of data groups
n = grouped_df.size()[0] #population of data per group
N = k * n #total population of data
```

```python
#Degree of Freedom
DF_total = N - 1
DF_between = k - 1
DF_within = N - k
```

```python
# Calculate Group Mean (GM)
GM = grouped_df.Score.mean()
```

```python
# Calculate Population Mean (PM)
PM = medication_df.Score.mean()
```

```python
# Calculate Sum of Squares (SS)
SS_between = (((GM - PM) ** 2) * n).sum()
SS_within = sum(medication_df['Score'].values ** 2) - sum(grouped_df.sum()['Score']**2)/n
SS_total = ((medication_df['Score'].values - PM) ** 2).sum()
```

```python
# Calculate Mean Sum of Squares
MS_between = SS_between / DF_between
MS_within = SS_within / DF_within
```

```python
# Calculate F-Value
F_value = MS_between/MS_within
```

```python
# Calculate p_value
p_value = stats.f.sf(F_value, DF_between, DF_within)
```

```python
print(F_value,p_value)
```

## Interpretation:
The calculated value of F is 86.3333 which is significantly greater than the critical value of F i.e. 3.5546.
As F_value > F_critical_value, null hypothesis is rejected and alternative hypothesis is accepted.

## Conclusion:
The three different groups of medications used in the experiment do not have the same effect.
